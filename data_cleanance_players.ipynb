# Data Cleanance

```python
import os
import pandas as pd
```

Read Players CSV File

```python
df = pd.read_csv("data/players.csv")
```

Filter for 21th century Men's World Cup

```python
df = df[df["female"] == 0]

df["tournaments_21c"] = df["list_tournaments"].apply(
    lambda s: [int(y.strip()) for y in s.split(",") if int(y.strip()) > 1999]
)
df = df[df["tournaments_21c"].map(len) > 0]

df["list_tournaments"] = df["tournaments_21c"]
df["count_tournaments"] = df["tournaments_21c"].map(len)
```

Merge player name columns to one and translate date of birth to age 

```python
df["player_name"] = df["given_name"].str.strip() + " " + df["family_name"].str.strip()
df.drop(columns=["given_name", "family_name"], inplace=True)

df["birth_year"] = pd.to_numeric(
    df["birth_date"].str[:4],
    errors="coerce"
)

df["age"] = (2026 - df["birth_year"]).astype("Int64")
```

Keep relevant columns

```python
df = df[[
    "player_id",
    "player_name",
    "birth_year",
    "age",
    "goal_keeper",
    "defender",
    "midfielder",
    "forward",
    "count_tournaments",
    "list_tournaments"
]]

df.rename(columns={
    "list_tournaments": "list_tournament_years"
}, inplace=True)

df.head()
```

Output file

```python
os.makedirs("output", exist_ok=True)

df.to_csv(
    "data/players_clean.csv",
    index=False
)
```
