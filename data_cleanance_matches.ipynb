# Data Cleanance

```python
import numpy as np
import os
import pandas as pd
```

Read Matches CSV File

```python
df = pd.read_csv("data/matches.csv")
```

Filter for 21th century Men's World Cup

```python
df = df[
    df["tournament_name"].str.contains("FIFA Men's World Cup", na=False)
]

df["match_date"] = pd.to_datetime(df["match_date"])
df = df[df["match_date"].dt.year > 1999]
```

Keep relevant columns

```python
df = df[[
    "match_id",
    "home_team_name",
    "away_team_name",
    "home_team_score",
    "away_team_score"
]]

df.columns = ["match_id", "team1", "team2", "team1_score", "team2_score"]
df.head()
```

Output file

```python
os.makedirs("output", exist_ok=True)

df.to_csv(
    "data/matches_clean.csv",
    index=False
)
```
