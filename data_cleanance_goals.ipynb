# Data Cleanance

Read Goals CSV File

```python
import os
import pandas as pd

df = pd.read_csv("data/goals.csv")
```

Filter for 21th century Men's World Cup

```python
df = df[
    df["tournament_name"].str.contains("FIFA Men's World Cup", na=False)
]

df["match_date"] = pd.to_datetime(df["match_date"])
df = df[df["match_date"].dt.year > 1999]
```

Merge player name columns to one

```python
df["player_name"] = df["given_name"].str.strip() + " " + df["family_name"].str.strip()
df.drop(columns=["given_name", "family_name"], inplace=True)
```

Add column for opponent team 

```python
teams = df["match_name"].str.split(" vs ", expand=True)
df["home_team"] = teams[0]
df["away_team"] = teams[1]

df["opponent_team"] = df.apply(
    lambda row: row["away_team"]
    if row["team_name"] == row["home_team"]
    else row["home_team"],
    axis=1
)

df.drop(columns=["match_name", "home_team", "away_team"], inplace=True)
```

Keep relevant columns

```python
df = df[[
    "goal_id",
    "match_id",
    "team_name",
    "opponent_team",
    "player_id",
    "player_name"
]]

df.rename(columns={
    "team_name": "scoring_team"
}, inplace=True)
df.head()
```

Output file

```python
os.makedirs("output", exist_ok=True)

df.to_csv(
    "data/goals_clean.csv",
    index=False
)
```
